# Resample

# Let's say we want to agg by Quarterly level

### rule can be found in https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#offset-aliases

```python
df.resample(rule='A').mean()
```

```python
df.resample('QS').sum()
```

```python
df.resample('Q').sum()
```

# Custom resampling methods

```python
def last_entry(entry):
    if len(entry)!= 0:
        return entry[-1]
```

```python
df.resample('A').apply(last_entry)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df['Volume'].resample("Q").mean().plot(kind='bar')
```

# Statsmodel Time Series API

# Hordick Prescot Filter to seperate out the trend and cyclic component of a given time series data

```python
from statsmodels.tsa.filters.hp_filter import hpfilter
```

```python
df = pd.read_csv(r"E:\Course\Python for Time Series Data Analysis\01. Introduction\UDEMY_TSA_FINAL\Data\macrodata.csv", parse_dates=True, index_col=0)
```

```python
df
```

```python
gdp_cyc, gdp_trend = hpfilter(df['realgdp'], lamb=1600) # as quarterly data
```

```python
df['realgdp'].plot()
gdp_trend.plot();
gdp_cyc.plot()
```
